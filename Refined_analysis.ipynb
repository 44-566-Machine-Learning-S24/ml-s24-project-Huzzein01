### Clustering 

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('MachineLearningProject1.csv')
X_clustering = data[['CarCount', 'BikeCount', 'BusCount', 'TruckCount']]

# Perform K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_clustering)

# Visualize clustering results
plt.scatter(X_clustering['CarCount'], X_clustering['BikeCount'], c=kmeans.labels_, cmap='viridis')
plt.xlabel('CarCount')
plt.ylabel('BikeCount')
plt.title('K-means Clustering')
plt.show()
```

### Dimension Reduction 

```python
from sklearn.decomposition import PCA

X_dim_reduction = data[['CarCount', 'BikeCount', 'BusCount', 'TruckCount']]

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_dim_reduction)

# Visualize PCA results
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA')
plt.show()
```

```python
plt.plot(data['Date'], data['CarCount'], label='CarCount')
plt.plot(data['Date'], data['BikeCount'], label='BikeCount')
plt.plot(data['Date'], data['BusCount'], label='BusCount')
plt.plot(data['Date'], data['TruckCount'], label='TruckCount')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Traffic Counts Over Time')
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import IsolationForest

X_anomaly_detection = data[['CarCount', 'BikeCount', 'BusCount', 'TruckCount']]

clf = IsolationForest(random_state=42)
clf.fit(X_anomaly_detection)
anomalies = clf.predict(X_anomaly_detection)

# Print anomalous data points
print("Anomalous data points:")
print(data[anomalies == -1])
```

```python
from sklearn.neural_network import MLPClassifier

X_classification = data[['CarCount', 'BikeCount', 'BusCount', 'TruckCount']]
y_classification = data['Traffic Situation']


mlp_classifier = MLPClassifier(max_iter=1000, random_state=42)
mlp_classifier.fit(X_classification, y_classification)
```

### Advanced Regression 

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPClassifier

X_regression = data[['CarCount', 'BikeCount', 'BusCount', 'TruckCount']]
y_regression = data['Total']

X_classification = data[['CarCount', 'BikeCount', 'BusCount', 'TruckCount']]
y_classification = data['Traffic Situation']

# Train Random Forest Regression
rf_regressor = RandomForestRegressor(random_state=42)
rf_regressor.fit(X_regression, y_regression)

# Train Multilayer Perceptron Classifier 
mlp_classifier = MLPClassifier(random_state=42)
mlp_classifier.fit(X_classification, y_classification)
```
